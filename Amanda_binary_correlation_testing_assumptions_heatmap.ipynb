```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

First, the index coloumn is dropped in the table below.

```python
table = pd.read_csv("data/polling_site_data/clean_precincts_with_polling_site.csv", index_col = 0)
df = table.drop(columns=['Total','Hispanic','White','Black','Asian','Mixed','Others'])
df.head()
```

Then, a heatmap is created to test whether there is multicollinearity between the independent variables.Correlation between the independent variables will range from 0 to 1, with 1 being the highest level of correlation.[(Leavitt, 2019)](https://https://towardsdatascience.com/multicollinearity-why-is-it-a-problem-398b010b77ac)



```python
plt.figure(figsize = (38,16))
sns.heatmap(df.corr(), annot = True)
plt.savefig('heatmap.png')
plt.show()
```
